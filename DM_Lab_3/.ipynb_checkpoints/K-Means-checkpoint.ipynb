```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = open(r'C:\Users\Domovoy\Downloads\points datasets-20201010T153653Z-001\points datasets\birch1.txt')
text = f.readlines()
data = np.array((len(text), 2))

data = np.array(([line.split() for line in text]), dtype=np.float64)
data
```

```python
plt.scatter(data[:,0], data[:,1], s=5, cmap='viridis');
```

```python
from sklearn.metrics import pairwise_distances_argmin

def find_clusters(X, n_clusters, rseed=2):
    # 1. РАНДОМНЫЙ ВЫБОР КЛАСТЕРОВ
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]
    
    while True:
        # 2a. ПРИСВОИТЬ ЛЭЙБЛЫ НА ОСНОВЕ БЛИЖАЙШЕГО ЦЕНТРА
        labels = pairwise_distances_argmin(X, centers)
        
        # 2b. НАЙТИ НОВЫЕ ЦЕНТРЫ НА ОСНОВЕ СРЕДНЕГО ТОЧЕК
        new_centers = np.array([X[labels == i].mean(0)
                                for i in range(n_clusters)])
        
        # 2c. ПРОВЕРКА НА СХОДИМОСТЬ
        if np.all(centers == new_centers):
            break
        centers = new_centers
    
    return centers, labels


centers, labels = find_clusters(data, 7)

plt.figure(figsize=(10,10))
plt.scatter(data[:, 0], data[:, 1], c=labels, s=2, cmap='viridis');
plt.scatter(centers[:,0], centers[:,1], c='black', s=50);
```

```python
for i, center in enumerate(centers):
    print(f'Center #{i+1}: x:{center[0]} | y:{center[1]}')
```
